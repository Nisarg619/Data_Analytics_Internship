```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('titanic')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
df.drop_duplicates(inplace=True)
print(df.isnull().sum())
```

```python
sns.heatmap(df.isnull())
plt.show()
sns.pairplot(df,hue='survived')
plt.show()
```

```python
print(df.groupby('class')['survived'].mean())
```
